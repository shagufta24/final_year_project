```python
members = ["a", "b", "c"]
indices = [5, 9, 2]
distances = [8, 3, 6]
```

```python
sorted_members = [member for dist, member in sorted(zip(distances, members), key=lambda x: x[0], reverse=True)]
sorted_members
```

```python
sorted_indices = [index for dist, index in sorted(zip(distances, indices), key=lambda x: x[0], reverse=True)]
sorted_indices
```

```python
import csv
with open('o1.csv', mode ='r') as f1: 
  o1 = list(csv.reader(f1))

with open('o2.csv', mode ='r') as f2: 
  o2 = list(csv.reader(f2))
```

```python
len(o1)
```

```python
from matplotlib import pyplot as plt

for m1, m2 in zip(o1, o2):
    plt.scatter(m1, m2)
    plt.xlabel("Function 1")
    plt.ylabel("Function 2")
    plt.title("Pareto front")
    plt.show()
```
